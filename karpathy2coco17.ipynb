```python
from pycocotools.coco import COCO
import json
import copy
from mmcv.utils import ProgressBar
```

```python
karpathy_split = json.load(open("/data/home/v-lixxue/coco17/karpathy_split/dataset_coco.json"))
```

```python
train_coco = COCO("/data/home/v-lixxue/coco17/annotations/captions_train2017.json")
val_coco = COCO("/data/home/v-lixxue/coco17/annotations/captions_val2017.json")
train_imgids = train_coco.getImgIds()
train_imgs = train_coco.loadImgs(train_imgids)
val_imgids = val_coco.getImgIds()
val_imgs = val_coco.loadImgs(val_imgids)
```

```python
prog_bar = ProgressBar(len(coco17_split['images']))
coco17_split = copy.deepcopy(karpathy_split)
for i in range(len(coco17_split['images'])):
    image = coco17_split['images'][i]
    if image['cocoid'] in train_imgids:
        image['filepath'] = 'train2017'
        coco17_image = train_coco.loadImgs(image['cocoid'])[0]
        image['filename'] = coco17_image['file_name']
        image['split'] = 'train'
    elif image['cocoid'] in val_imgids:
        image['filepath'] = 'val2017'
        coco17_image = val_coco.loadImgs(image['cocoid'])[0]
        image['filename'] = coco17_image['file_name']
        image['split'] = 'val'
    else:
        print("image id {} not found!".format(image['cocoid']))
        break
    prog_bar.update()
    
```
